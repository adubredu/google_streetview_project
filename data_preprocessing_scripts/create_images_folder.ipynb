```python
import os
import shutil
```

```python
stamford_not_found = []
for filename in os.listdir("Stamford/"):
    try:
        shutil.copy2('Stamford/'+filename+'/gsv_0.jpg', 'stamford_images_only/'+filename+'.jpg')
    except:
        print(filename, 'not found')
        stamford_not_found.append(filename)
        
```

```python
len(stamford_not_found)
```

```python
for folder in stamford_not_found:
    shutil.rmtree("Stamford/"+folder)
    print(folder, "removed")
```
