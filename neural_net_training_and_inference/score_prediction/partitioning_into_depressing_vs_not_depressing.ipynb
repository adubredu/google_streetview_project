```python
import numpy as np 
import pandas as pd 
```

```python
sf = pd.read_csv('stamford_dep_scores.csv')
nh = pd.read_csv('newhaven_dep_scores.csv')
hf = pd.read_csv('hartford_dep_scores.csv')
```

```python
scores = list(sf['depressing_score'])+list(nh['depressing_score'])+list(hf['depressing_score'])
mean = np.mean(scores)
std = np.std(scores)
print('mean is: {}'.format(mean))
print('std is: {}'.format(std))
```

```python
len(scores)
```

```python
sf_scores=[]
for score in sf['depressing_score']:
    if score < (mean-std):
        sf_scores.append(0) #not depressing
    elif score > (mean+std):
        sf_scores.append(1) #depressing
    else:
        sf_scores.append(2) #neutral
sf['score_labels']=sf_scores
sf.head(10)
```

```python
one,zero,two=0,0,0
for s in sf_scores:
    if s==1:
        one+=1
    elif s==0:
        zero+=1
    elif s==2:
        two+=1
print('no. of depressing: {}'.format(one))
print('no. of not depressing: {}'.format(zero))
print('no. of neutral: {}'.format(two))
```
